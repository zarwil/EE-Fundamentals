```
import matplotlib.pyplot as plt
plt.rcParams.update({'font.family':"sans serif",'mathtext.fontset':'cm'})
fig = plt.figure(figsize=([9,9]))
ax= fig.add_axes([0,0,1,1])
ax.axis('off')
ax.text(.5,.92,'RMS of Non-Sinusoidal Periodic Function $v(t)$',size=28,ha='center',va='center')
ax.text(.5,.82,r'$V_{RMS} = \sqrt{\frac{1}{T}\int_0^T v^2(t) dt} $',
        size=25,ha='center',va='top',multialignment='right',
         bbox=dict(boxstyle='ellipse',pad=0,facecolor='w',edgecolor='tab:orange',lw=5))
ax.text(.5,.65,'Fourier\nSeries',c='tab:blue',size=20,bbox=dict(boxstyle='larrow',pad=.1,facecolor='w',edgecolor='tab:orange',lw=3),
        ha='center',va='top',multialignment='center',rotation=90)

ax.text(.5,.4,r'$V_{RMS} = \sqrt{\frac{1}{T}\int_0^T \left[V_0+\sum_{k=1}^{\infty}V_k\cos(k\omega t+\theta_k) \right]^2 dt}$',
        size=25,ha='center',va='center',multialignment='right',
        bbox=dict(boxstyle='round',pad=.5,facecolor='w',edgecolor='tab:blue',lw=3))
ax.text(.5,.225,r'$V_{RMS}=\sqrt{V_0^2 + \frac{1}{2}\sum_{k=1}^{\infty}V_k^2}=\sqrt{V_0^2 + \sum_{k=1}^{\infty}V_{k\_RMS}^2}$',
        size=30,ha='center',va='top',multialignment='center',
         bbox=dict(boxstyle='roundtooth',pad=1,facecolor='w',edgecolor='tab:green',lw=6))
fig.savefig('rms_NonSineFunction.jpg',dpi=600)
# fig.savefig('rms_sineFunction.pdf',dpi=600)
```
