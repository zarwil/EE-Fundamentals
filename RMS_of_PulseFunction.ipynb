```
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams.update({'font.family':"sans serif",'mathtext.fontset':'cm'})
fig = plt.figure(figsize=([9,9]))
ax= fig.add_axes([0,0,1,1],xlim=[0,1],ylim=[0,1])
ax.axis('off')
x0,y0=.1,.6
xn,xm,yn,ym=x0-.02,.9,y0-.02,.86
[ax.annotate('',[(xm,y0),(x0,ym)][k],[(xn,y0),(x0,yn)][k],arrowprops=dict(arrowstyle='->')) for k in range(2)]
[ax.text([x0,xm][k],[ym,y0][k],['$\ v(t)$','$t$ '][k], size=25,
         va=['center','top'][k],ha=['left','right'][k]) for k in range(2)]
t=np.linspace(0,1,2000)
vt = np.heaviside(.6-1/np.pi*np.arccos(np.cos(4*np.pi*(t-.6/4))),0)
ax.plot(x0+t*(xm-x0)*.95,y0+vt*(ym-y0)*.8,lw=4,c='tab:orange')
ax.text(x0,y0+(ym-y0)*.8,'$V_m$ ',size=20,ha='right',va='center')
[ax.text(x0+(xm-x0)*.95*[.3,.5][k],y0,['$DT$','$T$'][k],size=20,ha='center',va='top') for k in range(2)]
ax.annotate('',(x0,y0+(ym-y0)*.4),(x0+(xm-x0)*.95*.3,y0+(ym-y0)*.4),arrowprops=dict(arrowstyle='<->'))
ax.text(x0+(xm-x0)*.95*.15,y0+(ym-y0)*.4+.01,'$T_p$',size=20,ha='center')

ax.text(.5,.92,r'RMS of Pulse Function of Duty Ratio $D=\frac{T_p}{T}$',size=30,ha='center',va='center')
ax.text(.5,.425, r'$V_{RMS} = \sqrt{\frac{1}{T}\int_0^T v^2(t) dt} = \sqrt{\frac{1}{T}\int_0^{DT} V_m^2 dt}$',
        size=25,ha='center',va='center',multialignment='right',
        bbox=dict(boxstyle='round',pad=.5,facecolor='w',edgecolor='tab:blue',lw=3))
ax.text(.5,.2,r'$V_{RMS}=\sqrt{D}V_m$',size=35,ha='center',va='top',multialignment='center',
         bbox=dict(boxstyle='roundtooth',pad=1,facecolor='w',edgecolor='tab:green',lw=6))
fig.savefig('rms_PulseFunction.jpg',dpi=600)
# fig.savefig('rms_sineFunction.pdf',dpi=600)
```
